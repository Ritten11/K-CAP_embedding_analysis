```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Create a 2x2 grid for subplots with different sizes
fig = plt.figure(figsize=(10, 8))

# Define the grid layout using GridSpec
gs = GridSpec(3, 2, width_ratios=[1, 1], height_ratios=[2, 1, 1])

# Subplot 1 (Top-left)
ax1 = plt.subplot(gs[0:1, 0])
ax1.plot([0, 1], [0, 1])
ax1.set_title("Subplot 1")

# Subplot 2 (Top-right)
ax2 = plt.subplot(gs[0, 1])
ax2.plot([0, 1], [0, 1], color='orange')
ax2.set_title("Subplot 2")

# Subplot 3 (Bottom-left)
ax3 = plt.subplot(gs[2, 0])
ax3.plot([0, 1], [0, 1], color='green')
ax3.set_title("Subplot 3")

# Subplot 4 (Bottom-right)
ax4 = plt.subplot(gs[1:2, 1])
ax4.plot([0, 1], [0, 1], color='red')
ax4.set_title("Subplot 4")

# Adjust layout to avoid overlapping titles and labels
# plt.tight_layout()

# Show the plot
# plt.show()
```
